```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python
#carga de un archivo excel
Pretest= pd.read_excel("PRETEST - Conocimiento STEM pedagógico tecnológico para el diseño de lecciones educativas.xlsx")
Pretest
```

```python
#Verificamos información del DataFrame
Pretest.info()
```

```python
#Obtengo un análisis univariado de una variable categórica en específico
table1= freq_tbl(Pretest['¿Cuál considera que es su nivel de competencia tecnológica?'])
table1
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro= table1[table1['frequency']>1]
Filtro
```

```python

#Ajusto el indice de mi dataframe
Filtro_index= Filtro.set_index('¿Ha participado en procesos relacionados con la educación STEM (Ciencia, Tecnología, Ingeniería y Matemáticas) ?:')
Filtro_index
```

```python
#Elimino las columnas de "percentage" y "cumulative_perc"
Filtro_index_frequency= Filtro_index.drop(['percentage','cumulative_perc'], axis=1)
Filtro_index_frequency
```

```python
#Realizamos grafico de barras del dataframe filtrado
Filtro_index_frequency.plot(kind = 'bar', width=0.8, figsize=(8,4), color= 'blue')
plt.title('¿Ha participado en procesos relacionados con la educación STEM?')
plt.xlabel('Tipo de Herramienta')
plt.ylabel('Frecuencia')
```

```python
#Realizamos grafico de pastel del dataframe filtrado
Filtro_index["frequency"].plot(kind='pie', figsize=(10,5), shadow=True, autopct="%0.1f%%")
plt.title('¿Ha participado en procesos relacionados con la educación STEM?')
```
